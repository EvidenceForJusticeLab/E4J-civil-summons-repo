### Checking why there are no September dates
- Going through each individual member 2021 ticket numbers
- Cross-referencing with both pulled nypd oath files for september dates

```python
import pandas as pd
import os
os.getcwd()
```

```python
oath_1 = pd.read_csv("/Users/joyamelvin/Desktop/Evidence for Justice Lab/officer_dem/OATH_Hearings_Division_Case_Status_NYPD ONLY (Jan 1 2021 to Aug 31 2022).csv", low_memory=False)

oath_2 = pd.read_csv("/Users/joyamelvin/Desktop/Evidence for Justice Lab/officer_dem/OATH_Hearings_Division_Case_Status_20260109.csv", low_memory=False)


```

```python
# Load September NYPD tickets and clean
oath1_sept = oath_1[oath_1['Violation Date'].astype(str).str.startswith('9/')]['Ticket Number'].astype(str).str.strip()
print(f" OATH 1 September NYPD ticket numbers: {len(oath1_sept)}")
```

```python
# Load September NYPD tickets and clean
oath2_sept = oath_2[oath_2['Violation Date'].astype(str).str.startswith('9/')]['Ticket Number'].astype(str).str.strip()
print(f"OATH 2 September NYPD ticket numbers: {len(oath2_sept)}")
```

```python
files = [
    '/Users/joyamelvin/Desktop/Evidence for Justice Lab/2021_RA_Citation_Files/2021 Civil Citation Data Entry Template JWM_03.20.25.csv',
    '/Users/joyamelvin/Desktop/Evidence for Justice Lab/2021_RA_Citation_Files/2021 Civil Citation Data Entry Bagley_2025.02.23.csv',
    '/Users/joyamelvin/Desktop/Evidence for Justice Lab/2021_RA_Citation_Files/2021 Civil Citation Data Entry Template AG 2025.02.27.csv',
    '/Users/joyamelvin/Desktop/Evidence for Justice Lab/2021_RA_Citation_Files/2021 Civil Citation Data Entry Template_Baugh_2025.03.07.csv',
    '/Users/joyamelvin/Desktop/Evidence for Justice Lab/2021_RA_Citation_Files/2021 Civil Citation Data Entry Template_Pozo_2025.03.08.csv',
    '/Users/joyamelvin/Desktop/Evidence for Justice Lab/2021_RA_Citation_Files/2021 Civil Citation Data Entry Template Johnson_2025.01.30.csv'
]

# Load and combine all ticket numbers
combined = pd.concat([
    pd.read_csv(f, skiprows=1, dtype=str)[['Ticket Number ']].rename(columns={'Ticket Number ': 'Ticket_Number'})
    for f in files
], ignore_index=True)

# Clean ticket numbers
combined['Ticket_Number'] = combined['Ticket_Number'].str.strip().str.lstrip('0')
combined = combined.dropna(subset=['Ticket_Number'])

print(f"Total tickets in combined (with duplicates): {len(combined)}")
print(f"Unique tickets in combined: {combined['Ticket_Number'].nunique()}")
```

```python
# Compare OATH 1 match to lab members 
overlap = set(combined['Ticket_Number']).intersection(set(oath1_sept))
print(f"\nSeptember NYPD tickets: {len(oath1_sept)}")
print(f"Combined tickets that match September NYPD: {len(overlap)}")
print(f"September tickets NOT in combined: {len(set(oath1_sept)) - len(overlap)}")
```

```python
# Compare OATH 2 match to lab members 
overlap = set(combined['Ticket_Number']).intersection(set(oath2_sept))
print(f"\nSeptember NYPD tickets: {len(oath2_sept)}")
print(f"Combined tickets that match September NYPD: {len(overlap)}")
print(f"September tickets NOT in combined: {len(set(oath2_sept)) - len(overlap)}")
```
