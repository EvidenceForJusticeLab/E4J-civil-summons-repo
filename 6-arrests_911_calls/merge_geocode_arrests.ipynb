#### Merge the Geocoded Citations with Geocoded Arrests and Calls for Service

```python
##setup
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
##import citations and arrests
citations = pd.read_csv('./Geocoding/data/civil_summons_2021_v6.csv')
arrests = pd.read_csv('./911_arrest /data/output/arrests_by_census_block_month_year.csv')
```

```python
##import calls for service
calls = pd.read_csv('./911_arrest /data/output/final_calls_by_block_month_year.csv')
```

```python
##check citations file
citations.head()
```

```python
##convert the census level data to a string 
citations["BCTCB2020"] = citations["BCTCB2020"].astype(str)
```

```python
##check arrests file
arrests.head()
```

```python
##convert census level data to a string
arrests["BCTCB2020"] = arrests["BCTCB2020"].astype(str)
```

```python
##filter to only include arrests from 2021
arrests_2021 = arrests[arrests["year"] == 2021]
```

```python
##check calls for service file
calls.head()
```

```python
##convert census level data to a string
calls["BCTCB2020"] = calls["BCTCB2020"].astype(str)
```

```python
##merge arrests w/ citations
citation_arrests_bctcb = citations.merge(arrests_2021, left_on=["month_issued", "BCTCB2020"], right_on=["month", "BCTCB2020"], how="left")
citation_arrests_bctcb.head()
```

```python
##clean up column names
citation_arrests_bctcb.drop(columns=['year', 'month', 'CB2020_y', 'CT2020_y'], inplace=True)
citation_arrests_bctcb.rename(columns={'CB2020_x': 'CB2020', 'CT2020_x' : 'CT2020'}, inplace=True)
```

```python
##clean NAs
citation_arrests_bctcb.loc[citation_arrests_bctcb['total_arrests'].isna(), 'total_arrests'] = 0
```

```python
##check the merged file
citation_arrests_bctcb.head(25)
```

```python
##merge the citation/arrest file with calls for service by census data
citation_arrests_calls = citation_arrests_bctcb.merge(calls, left_on=["month_issued", "BCTCB2020"], right_on=["month", "BCTCB2020"], how="left")
citation_arrests_calls.head()
```

```python
##clean up column names
citation_arrests_calls.drop(columns=['year', 'month', 'CB2020_y', 'CT2020_y'], inplace=True)
citation_arrests_calls.rename(columns={'CB2020_x': 'CB2020', 'CT2020_x' : 'CT2020'}, inplace=True)
```

```python
##get the total number of rows with no calls for service
citation_arrests_calls['total_calls'].isna().sum()
```

```python
##clean NAs
citation_arrests_calls.loc[citation_arrests_calls['total_calls'].isna(), 'total_calls'] = 0
```

```python
##export to csv
citation_arrests_calls.to_csv('./civil_summons_2021_v7.csv', index = False)
```
